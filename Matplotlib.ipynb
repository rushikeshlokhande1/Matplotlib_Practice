```python
import matplotlib.pyplot as plt

plt.figure(figsize=(9,5))

# Line plot
plt.plot(df["date"], df["sales"], label="Sales", linewidth=2)

# Find max & min
max_sales = df["sales"].max()
min_sales = df["sales"].min()

max_date = df.loc[df["sales"] == max_sales, "date"].values[0]
min_date = df.loc[df["sales"] == min_sales, "date"].values[0]

# Highlight points
plt.scatter(max_date, max_sales, marker='o', s=100, label="Max Sales")
plt.scatter(min_date, min_sales, marker='o', s=100, label="Min Sales")

# Annotations
plt.annotate(f"Max: {max_sales}",
             (max_date, max_sales),
             textcoords="offset points",
             xytext=(0,10),
             ha='center')

plt.annotate(f"Min: {min_sales}",
             (min_date, min_sales),
             textcoords="offset points",
             xytext=(0,-15),
             ha='center')

plt.title("Sales Trend with Max & Min Highlighted")
plt.xlabel("Date")
plt.ylabel("Sales")

plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

# Scatter plot
plt.scatter(df["advertising_spend"], df["sales"], marker='o')

# Regression line
x = df["advertising_spend"]
y = df["sales"]

m, b = np.polyfit(x, y, 1)   # slope & intercept
plt.plot(x, m*x + b)

# Labels & title
plt.title("Correlation Analysis")
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")

plt.grid(True)
plt.show()
```
